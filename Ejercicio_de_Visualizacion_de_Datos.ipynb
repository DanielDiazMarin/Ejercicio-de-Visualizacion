```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
filename = '/content/Cost of 1GB of Data.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
plt.figure(figsize=(30, 6))
plt.style.use('seaborn-pastel')
sns.scatterplot(data=df, x = 'Country', y = 'Avg Price of 1GB (USD)', hue = 'Country', legend = False)
plt.xticks(rotation = 90)
plt.ylim(bottom=0, top=30)
plt.xlabel('Country', fontsize = 15)
plt.ylabel('Avg Price of 1GB', fontsize = 15)
plt.grid()
plt.title('Average Price of 1GB (USD)', fontsize = 15)
plt.show()
```
